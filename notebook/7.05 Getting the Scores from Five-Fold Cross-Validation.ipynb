```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
file_url = r'https://raw.githubusercontent.com/sedeba19/Chapter-7/main/data/Dataset_car.data.txt'
```

```python
# col_names
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'car_class']
# Load the dataset
df = pd.read_csv(file_url,
                 names = col_names, 
                 header=None)
df.head()
```

```python
df.info()
```

```python
# Split the data ito 80% training and 20% testing set
training_df, eval_df = train_test_split(df, test_size=0.8, random_state=0)
```

```python
training_df.info()
```

```python
eval_df.info()
```

```python
# Encode categorical variables
df_dumm = pd.get_dummies(df, 
                         columns=['buying', 
                                  'maint', 
                                  'doors', 
                                  'persons', 
                                  'lug_boot', 
                                  'safety'])
df_dumm
```

```python
# Separate features and labels DataFrames
features = df_dumm.drop(['car_class'], axis=1).values
labels = df_dumm[['car_class']].values
```

```python
from sklearn.linear_model import LogisticRegression
# Create an instance of LogisticRegression
lr_model = LogisticRegression()

from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr_model, 
                         features, 
                         labels, 
                         cv=5)
print(scores)
```
